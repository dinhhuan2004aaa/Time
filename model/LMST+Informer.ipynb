```
data = df[['TMP_mean','TMP_max','TMP_min','DEW_mean','DEW_max','DEW_min']].values
```

```
from sklearn.preprocessing import MinMaxScaler
```

```
# Chuẩn hóa dữ liệu (MinMaxScaler)
scaler = MinMaxScaler()
# Lấy 70% đầu làm tập train
n_total = len(data)
n_train = int(n_total * 0.7)
train_data = data[:n_train]
scaler.fit(train_data)           # chỉ fit trên train
data_scaled = scaler.transform(data)
```

```
# Tạo chuỗi 30 ngày -> dự đoán 7 ngày kế
X, y = [], []
n_input = 96
n_output = 24
for i in range(n_total - n_input - n_output + 1):
    X.append(data_scaled[i:i+n_input, :])           # 6 biến
    y.append(data_scaled[i+n_input:i+n_input+n_output, 0])  # chỉ TMP_mean tại cột 0
X = np.array(X)  # shape (samples,30,6)
y = np.array(y)  # shape (samples,7)
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model = Sequential()
model.add(LSTM(50, activation='tanh', input_shape=(n_input, 6)))
model.add(Dense(n_output))   # n_output = 7

model.compile(optimizer='adam', loss='mse')
model.summary()
```

```
# Chia X, y thành train/val/test theo chỉ số
n_val = int(len(X) * 0.1)
n_test = len(X) - n_train - n_val
X_train = X[:n_train]; y_train = y[:n_train]
X_val = X[n_train:n_train+n_val]; y_val = y[n_train:n_train+n_val]
X_test = X[n_train+n_val:]; y_test = y[n_train+n_val:]

history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=16,
    validation_data=(X_val, y_val)
)
```

```
y_pred_scaled = model.predict(X_test)
```

```
y_pred_tmp = y_pred_scaled.flatten()  # (699*7,) hoặc (samples*7,)
y_true_tmp = y_test.flatten()

# Tạo mảng đủ 6 cột để inverse
y_pred_full = np.zeros((len(y_pred_tmp), 6))
y_pred_full[:, 0] = y_pred_tmp  # TMP_mean ở cột 0

y_true_full = np.zeros((len(y_true_tmp), 6))
y_true_full[:, 0] = y_true_tmp

# Áp dụng inverse_transform và lấy lại TMP_mean
y_pred_inv = scaler.inverse_transform(y_pred_full)[:, 0]
y_true_inv = scaler.inverse_transform(y_true_full)[:, 0]
```

```
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
rmse = np.sqrt(mean_squared_error(y_true_inv, y_pred_inv))
mae = mean_absolute_error(y_true_inv, y_pred_inv)
print(f"Test RMSE: {rmse:.3f}, MAE: {mae:.3f}")
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training và Validation Loss theo từng Epoch')
plt.legend()
plt.grid(True)
plt.show()
```

```
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```
# Lấy các cột cần thiết và chuẩn hóa dựa trên tập huấn luyện
features = ['TMP_mean','TMP_max','TMP_min','DEW_mean','DEW_max','DEW_min']
train_cut = int(len(df)*0.8)
scaler = StandardScaler()
train_data = scaler.fit_transform(df[features].iloc[:train_cut])
val_data   = scaler.transform(df[features].iloc[train_cut:int(len(df)*0.9)])
test_data  = scaler.transform(df[features].iloc[int(len(df)*0.9):])
```

```
import numpy as np
def create_sequences(data, input_len=30, output_len=7):
    X, Y = [], []
    for i in range(len(data) - input_len - output_len + 1):
        X.append(data[i:i+input_len])            # 96 x 6 features
        Y.append(data[i+input_len:i+input_len+output_len, 0])  # 24 x TMP_mean
    return np.array(X), np.array(Y)

X_train, Y_train = create_sequences(train_data)
X_val,   Y_val   = create_sequences(val_data)
X_test,  Y_test  = create_sequences(test_data)
```

```
import torch
from torch import nn

class SimpleInformer(nn.Module):
    def __init__(self, input_dim=6, d_model=64, nhead=4, num_layers=2, output_len=7):
        super().__init__()
        # Mô đun embedding đầu vào
        self.proj_in = nn.Linear(input_dim, d_model)
        # Tọa độ vị trí (sinusoidal)
        pe = torch.zeros(1000, d_model)
        position = torch.arange(0, 1000).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2) * (-np.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        self.pos_encoder = pe  # (1000, d_model)
        # Transformer Encoder
        encoder_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=nhead)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        # Đầu ra tuyến tính cho 7 giá trị dự đoán
        self.fc_out = nn.Linear(d_model, output_len)

    def forward(self, x):
        # x: (batch, seq_len, features)
        x = x.permute(1, 0, 2)             # -> (seq_len, batch, features)
        x = self.proj_in(x)               # -> (seq_len, batch, d_model)
        # Thêm positional encoding
        seq_len = x.size(0)
        pe = self.pos_encoder[:seq_len, :].unsqueeze(1)  # (seq_len, 1, d_model)
        x = x + pe
        # Encoder
        x_enc = self.transformer_encoder(x)  # (seq_len, batch, d_model)
        # Dự báo: lấy bước thời gian cuối cùng
        out = x_enc[-1]                      # (batch, d_model)
        out = self.fc_out(out)              # (batch, 7)
        return out
```

```
from torch.utils.data import DataLoader, TensorDataset
import torch.optim as optim

# Chuyển dữ liệu sang Tensor và tạo DataLoader
train_ds = TensorDataset(torch.tensor(X_train).float(), torch.tensor(Y_train).float())
val_ds   = TensorDataset(torch.tensor(X_val).float(),   torch.tensor(Y_val).float())
test_ds  = TensorDataset(torch.tensor(X_test).float(),  torch.tensor(Y_test).float())
train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
val_loader   = DataLoader(val_ds, batch_size=32)
test_loader  = DataLoader(test_ds, batch_size=32)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = SimpleInformer().to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

best_val_loss = float('inf')
for epoch in range(20):
    model.train()
    for X_batch, Y_batch in train_loader:
        X_batch = X_batch.to(device)
        Y_batch = Y_batch.to(device)
        preds = model(X_batch)
        loss = criterion(preds, Y_batch)
        optimizer.zero_grad(); loss.backward(); optimizer.step()
    # Đánh giá trên tập xác thực
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for X_batch, Y_batch in val_loader:
            X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)
            preds = model(X_batch)
            val_loss += criterion(preds, Y_batch).item() * X_batch.size(0)
    val_loss /= len(val_ds)
    print(f"Epoch {epoch}: Val MSE={val_loss:.4f}")
    # Lưu mô hình tốt nhất
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), 'best_informer.pth')

# Đánh giá trên tập test
model.load_state_dict(torch.load('best_informer.pth'))
model.eval()
test_loss = 0
with torch.no_grad():
    for X_batch, Y_batch in test_loader:
        X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)
        preds = model(X_batch)
        test_loss += criterion(preds, Y_batch).item() * X_batch.size(0)
test_loss /= len(test_ds)
print(f"Test MSE = {test_loss:.4f}, RMSE = {test_loss**0.5:.4f}")
```

```
import matplotlib.pyplot as plt

# Tạo danh sách dự đoán trên tập test
model.eval()
all_preds = []
with torch.no_grad():
    for X_batch, _ in test_loader:
        X_batch = X_batch.to(device)
        preds = model(X_batch).cpu().numpy()
        all_preds.append(preds)
all_preds = np.vstack(all_preds)  # (num_samples, 7)
# Ta có thể chỉ lấy dự đoán ngày đầu tiên của mỗi cửa sổ để so sánh chuỗi
pred_series = all_preds[:, 0]
true_series = Y_test[:, 0]

# Vẽ kết quả
plt.figure(figsize=(8,4))
plt.plot(true_series, label='Thực tế')
plt.plot(pred_series, label='Dự đoán (Informer)')
plt.xlabel('Mẫu thử')
plt.ylabel('TMP\_mean')
plt.legend()
plt.show()
```
